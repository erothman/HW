In this notebook, we're going to have a closer look at Support Vector Machines.  Specifically, we'll look at the individual support vectors and the margins for each class.  We'll also compare with Logistic Regression.   In both cases, we'll use the sklearn implementations for simplicity.

**Some of the plots are a bit slow to generate, as they are training SVM classifiers.  They should appear after a short wait.**

```python
from data import load_data
from models import *
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from scipy.sparse import csr_matrix
np.random.seed(475)
```

For this part of the assignment, we'll generate a synthetic dataset from Gaussians. The code below generates a Gaussian with a given label, provided a mean, covariance matrix, and number of samples.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.patches import Rectangle

import matplotlib.style
import matplotlib as mpl
from sklearn.datasets import make_classification
%matplotlib inline
def generate_data(mean, covariance, num_samples, label):

    # Sample from a Gaussian
    X = np.random.multivariate_normal(mean, covariance, num_samples)
    
    if label == 1:
        y = np.ones(num_samples)
    else:
        y = np.zeros(num_samples)
    
    return X, y
```

## SVMs on Linearly Separable Data

Let's look at how SVMs handle data that is linearly separable.  

The first plot shows the data.  The SVM decision boundary is a solid line, the SVM margins are dashed lines, and the starred data points are the support vectors.  As shown in the first line, about 2% of the data points are support vectors.

In the second plot, we see a histogram of the absolute margin of each data point (on the X axis) compared to the counts of each margin (on the Y axis) for both Logistic Regression and SVM.  We can see that the decision function for SVM usually returns a smaller margin than Logistic Regression.

```python

def plot_stuff(model):
    plt.figure(figsize=(12,8))
    ax = plt.gca()
    y_color_dict = {0:"r", 1:"b"}
    y_colors = [y_color_dict[v] for v in y]
    ax.scatter(X[:, 0], X[:, 1],
                c = y_colors, alpha = .4)
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = model.decision_function(xy).reshape(XX.shape)
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    sv_array = model.support_vectors_.toarray()
    for i in range (sv_array.shape[0]):
        ylab0 = np.where((X[:,0] == sv_array[i,0]) & (X[:,1] == sv_array[i,1]))
        ax.scatter(sv_array[i,0], sv_array[i,1], s=100, marker='*', c=y_colors[ylab0[0][0]])
    perc_sv = float(sv_array.shape[0]) / float(X.shape[0])
    print("{0} support vectors, {1:0.2f}% of the input samples".format(sv_array.shape[0], perc_sv*100))
    plt.title("SVM margins and support vectors")
    plt.show()
    plt.clf()
    
    margins = model.decision_function(csr_matrix(X))
    
    lr_margins = lr_model1.decision_function(csr_matrix(X))
    hist_colors = ["m", 'y']
    plt.figure(figsize=(12,8))
    ax = plt.gca()
    ax.hist([np.absolute(lr_margins), np.absolute(margins)], bins=25, label=["LR", "SVM"], color=hist_colors)
    handles = [Rectangle((0,0),1,1,color=c,ec="k") for c in hist_colors]
    labels= ["LR", "SVM"]
    ax.legend(handles, labels)
    plt.title("Histogram of margins for Logistic Regression and SVM")
    plt.xlabel('Absolute margin')
    plt.ylabel('Number of training datapoints')
    
    plt.show()
    
X1, y1 = generate_data([-3,-3], [[1, -0.5],[-0.5, 1]], 1000, 1)
X2, y2 = generate_data([3,3], [[1, -0.5],[-0.5, 1]], 1000, 0)
X = np.vstack((X1, X2))
y = np.concatenate((y1, y2))

lr_model1 = LogisticRegression()
lr_model1.fit(csr_matrix(X), y)
sv_model0 = SVC(kernel='linear')
sv_model0.fit(csr_matrix(X), y)
plot_stuff(sv_model0)
```

## Data without a clear separating margin

Since the data had a clear margin separating the two classes, the SVM gave us a clear decision boundary with a small number of support vectors.

Let's create two Gaussians that don't have a clear margin of separation.  Also, let's fit a logistic regression model to the data to give up a point of comparison.

```python
X1, y1 = generate_data([-1.25,-1.25], [[1, -0.5],[-0.5, 1]], 1000, 1)
X2, y2 = generate_data([1.25,1.25], [[1, -0.5],[-0.5, 1]], 1000, 0)
X = np.vstack((X1, X2))
y = np.concatenate((y1, y2))

lr_model1 = LogisticRegression()
lr_model1.fit(csr_matrix(X), y)

plt.figure(figsize=(12,8))
ax = plt.gca()

y_color_dict = {0:"r", 1:"b"}
y_colors = [y_color_dict[v] for v in y]
ax.scatter(X[:, 0], X[:, 1],
            c = y_colors, alpha = .4)

xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = lr_model1.decision_function(xy).reshape(XX.shape)
ax.contour(XX, YY, Z, colors='k', levels=[0], alpha=0.5,
           linestyles=['-'])
```

## Question 1: SVMs on non-linearly separable data.


Let's see how SVMs do on this dataset.  We'll use fit a SVM model with a linear kernel and the default C setting (1).

**Notebook Question 1: Look at the histogram of margins for Logistic Regression and SVMs.  Describe the difference in margin distribution between the two models, and explain why this occurs. Answer this question in the corresponding box in the LaTeX template file.**

```python


svm_model1 = SVC(kernel='linear', C=1)
svm_model1.fit(csr_matrix(X), y)

plt.figure(figsize=(12,8))
ax = plt.gca()

ax.scatter(X[:, 0], X[:, 1],
            c = y_colors, alpha = .4)
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = svm_model1.decision_function(xy).reshape(XX.shape)
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])

sv_array = svm_model1.support_vectors_.toarray()
for i in range (sv_array.shape[0]):
    ylab0 = np.where((X[:,0] == sv_array[i,0]) & (X[:,1] == sv_array[i,1]))
    ax.scatter(sv_array[i,0], sv_array[i,1], s=100, marker='*', c=y_colors[ylab0[0][0]])
perc_sv = float(sv_array.shape[0]) / float(X.shape[0])
print("{0} support vectors, {1:0.2f}% of the input samples".format(sv_array.shape[0], perc_sv*100))
plt.title("SVM margins and support vectors")
plt.show()
plt.clf()

margins = svm_model1.decision_function(csr_matrix(X))

lr_margins = lr_model1.decision_function(csr_matrix(X))
hist_colors = ["m", 'y']
plt.figure(figsize=(12,8))
ax = plt.gca()
ax.hist([np.absolute(lr_margins), np.absolute(margins)], bins=25, label=["LR", "SVM"], color=hist_colors)
handles = [Rectangle((0,0),1,1,color=c,ec="k") for c in hist_colors]
labels= ["LR", "SVM"]
ax.legend(handles, labels)
plt.title("Histogram of margins for Logistic Regression and SVM")
plt.xlabel('Absolute margin')
plt.ylabel('Number of training datapoints')

plt.show()
```
